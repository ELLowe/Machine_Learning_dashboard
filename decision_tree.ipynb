```python
# Dependencies
from sklearn import tree
import pandas as pd
import os 
from sklearn.model_selection import train_test_split
```

```python
# Load in data
df = pd.read_csv('cleaned_disorder.csv')
df.head()
```

```python
# split into inputs and outputs
outcomes = df['disorder']
outcomes.head()

input_factors = df.drop(['disorder'], axis = 1)
input_factors.head()
```

```python
# Train/test Split
input_train, input_test, output_train, youtput_test = train_test_split(input_factors, outcomes, random_state=42)
```

```python
clf = tree.DecisionTreeClassifier()
clf = clf.fit(input_train, output_train)
clf.score(input_test, output_test)
```
