```python
# dependencies
import pandas as pd
import numpy
import os
from sklearn import tree
from sklearn.preprocessing import LabelEncoder
```

```python
# import CSV

df = pd.read_csv('cleaned_anxiety_testing_data.csv')
df.head()
```

```python
df = df.drop(['Subject','GAD Probabiliy - Gamma','SAD Probability - Gamma','Sample Weight'], axis = 1)
df.head()
```

```python
# add disorder column
df['disorder'] = 'place holder'
df.head()
```

```python
# create disorder function
def disorder_condenser(x):
    if x.SAD == 0 and x.GAD == 0:
        return 'None'
    return 'Anxiety'
```

```python
#apply funct to set
df['disorder'] = df.apply(disorder_condenser, axis=1)
```

```python
# drop gad/sad
df = df.drop(['GAD','SAD'], axis=1)
df.head()
```

```python
# Convert sex to binary int
df.Sex[df.Sex == 'M'] = 0
df.Sex[df.Sex == 'F'] = 1
df.head()
```

```python
# save updated DF to CSV
df.to_csv('cleaned_binary_disorder.csv')
```
