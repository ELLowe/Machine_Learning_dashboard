```python
# dependencies
import pandas as pd
import numpy
import os
from sklearn import tree
from sklearn.preprocessing import LabelEncoder
```

```python
# import CSV

df = pd.read_csv('cleaned_anxiety_testing_data.csv')
df.head()
```

```python
df = df.drop(['Subject','GAD Probabiliy - Gamma','SAD Probability - Gamma','Sample Weight'], axis = 1)
df.head()
```

```python
# add disorder column
df['disorder'] = 'place holder'
df.head()
```

```python
# create disorder function
def disorder_condenser(x):
    if x.SAD == 0 and x.GAD == 0:
        return 'None'
    elif x.SAD == 1 and x.GAD == 0:
        return 'SAD'
    elif x.SAD == 0 and x.GAD == 1:
        return 'GAD'
    return 'Both'
```

```python
#apply funct to set
df['disorder'] = df.apply(disorder_condenser, axis=1)
```

```python
# drop gad/sad
df = df.drop(['GAD','SAD'], axis=1)
df.head()
```

```python
# Convert sex to binary int
df.Sex[df.Sex == 'M'] = 0
df.Sex[df.Sex == 'F'] = 1
df.head()
```

```python
# save updated DF to CSV
df.to_csv('cleaned_disorder.csv')
```

```python
#divide data into inputs(associated factors) and outputs(disorder diagnoses)
data = df.values
inputs = data[:, 0:21]
outputs = data[:, 21]
```

```python
#  Label-encode data set
label_encoder = LabelEncoder()
label_encoder.fit(outputs)
encoded_outputs = label_encoder.transform(outputs)
```

```python
for label, original_class in zip(encoded_outputs, outputs):
    print('Disorder: ' + str(original_class))
    print('Encoded Label: ' + str(label))
    print('-' * 12)
```

```python
from keras.utils import to_categorical

# Step 2: One-hot encoding
one_hot_outputs = to_categorical(encoded_outputs)
one_hot_outputs
```
